```python
import pandas as pd
df=pd.read_csv("C:/Users/User/Downloads/Best_Games_of_All_Time.csv")
```

```python
df.info()
```

```python
# missing qiymatlarni to'ldirish
def tozalash(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
df=tozalash(df)
```
