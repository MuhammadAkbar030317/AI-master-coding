```python
import pandas as pd
df=pd.read_csv('C:/Users/User/Downloads/multilingual_mobile_app_reviews_2025.csv')
```

```python
df.info()
```

```python
def tozalash(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

```python
df=tozalash(df)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
def encodla(df):
    encoder=LabelEncoder()
    for col in df.columns:
        if df[col].dtype=='object':
            if df[col].nunique()<5:
                dummies=pd.get_dummies(df[col],prefix=col,dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encodla(df)
```

```python
# Scaling
from sklearn.preprocessing import MinMaxScaler
def scaling_qil(df):
    scaler=MinMaxScaler()
    num_col=df.select_dtypes(include=['int32','int64']).columns
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scaling_qil(df)
```

## Train test ga ajratishni boshlash

```python
x=df.drop("rating", axis=1).astype(int)
```

```python
y=df["rating"].astype(int)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
d_class=DecisionTreeClassifier()
```

```python
d_class.fit(x_train, y_train)
```

```python
y_pred=d_class.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
clas_rep=classification_report(y_test,y_pred)
```

```python
print(clas_rep)
```
