```python
import pandas as pd
df=pd.read_csv("/Users/User/Downloads/final_dataset.csv")
```

```python
df.info()
```

```python
def tozlash(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=="object":
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(), inplace=True)
    return df
```

```python
df=tozlash(df)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
def encoding_qilish(df):
    encoder=LabelEncoder()
    df=df.copy()
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique()<5:
                dummies=pd.get_dummies(df[col], prefix=col, dtype=int)
                df=pd.concat([df.drop(columns=[col]),dummies],axis=1)
            else:
                df[col]=encoder.fit_transform(df[col])
    return df
```

```python
df=encoding_qilish(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
df.info()
```

```python
def scaling_qilish(df):
    num_col=df.select_dtypes(include=["float64","int64"]).columns.drop("AQI")
    df[num_col]=scaler.fit_transform(df[num_col])
    return df
```

```python
df=scaling_qilish(df)
```

```python
from sklearn.linear_model import LogisticRegression
logic_reg=LogisticRegression()
```

```python
x=df.drop("AQI",axis=1)
```

```python
y=df["AQI"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split( x, y, test_size=0.2, random_state=42)
```

```python
logic_reg
```

```python
logic_reg.fit(x_train,y_train)
```

```python
y_predict=logic_reg.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
class_report=classification_report(y_test, y_predict)
```
