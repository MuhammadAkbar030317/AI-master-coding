```python
import pandas as pd
df=pd.read_csv("C:/Users/User/Desktop/AI/Datasets/steam_sales.csv")
```

```python
df.info()
```

```python
# Game Name ---------------------------- Title of the game on Steam.
# Rating ------------------------------- Average Steam user rating (out of 10).
# Reviews------------------------------- Total number of reviews submitted by users.
# Discount ----------------------------- Current discount percentage applied to the game.
# Price -------------------------------- Current discounted price in euros.
# Original Price ----------------------- Original price in euros before discount.
# Release Date-------------------------- Official release date of the game.
# Windows ------------------------------  1 if available on Windows, 0 otherwise.
# Linux--------------------------------- 1 if available on Linux, 0 otherwise.
# MacOS--------------------------------- 1 if available on macOS, 0 otherwise.
# Fetched At---------------------------- Timestamp of when the data was collected
```

```python
df["Rating"].value_counts()
```

```python
#  o'yin sifatini qandayligini aytuvchi feature yaratim
df["quality"]=df["Rating"].map({9.0:"very high", 10.0:"very high",8.0:"high", 7.0:"high", 5.0:"medium",6.0:"medium", 3.0:"low", 4.0:"low"})
```

```python
# necha pul discount olgani ni pulda hisoblaydigan ustun yaratdim
df["discount_in_price"]=(df["Original Price (€)"]*df["Discount%"])/100
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder , MinMaxScaler
class DataPreprocessing:
    def __init__(self, df):
        self.df=df
    def missing_values(self):
        for col in self.df.columns:
            if self.df[col].isnull().any():
                if self.df[col].dtype == "object":
                    self.df[col].fillna(self.df[col].mode()[0], inplace=True)
                else:
                    self.df[col].fillna(self.df[col].mean(), inplace=True)
        return self

    def encoding(self):
        encoder = LabelEncoder()
        for col in self.df.columns:
            if self.df[col].dtype == "object":
                if self.df[col].nunique() <= 5:
                    dummies = pd.get_dummies(self.df[col], dtype=int, prefix=col)
                    self.df = pd.concat((self.df.drop(columns=col), dummies), axis=1)
                else:
                    self.df[col] = encoder.fit_transform(self.df[col])
        return self
    
    def scaling(self):
        scaler=MinMaxScaler()
        for col in self.df.columns:
            num_col=self.df.select_dtypes(include=["float64","int64"]).columns.drop("Discount%")
            self.df[col]=scaler.fit_transform(self.df[num_col])
        return self

dp=DataPreprocessing(df)
dp.missing_values().encoding()
dp.scaling()
df=dp.df
```
